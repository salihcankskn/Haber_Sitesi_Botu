```python
import requests  # Web kazımada da kullandığımız bir method HTTP istekleri yapmak için kullanılır
from tkinter import Tk, Frame, Button, Label, Canvas, RAISED, TOP  # Tkinter modülünü ve arçlarını import ettik
import webview  # Linklere tıklandığında pythonun kendi içinde web sayfası açmasını sağlar.
import feedparser  # RSS leri pars etmek için kullanılır

#altın/gümüş fiyatlarını çekmek için kullandığımız url ve apı key  (goldapi adlı siteden aldım ancak TRY karşılığı yoktu ve dönüştüremedim)
API_KEY = 'goldapi-4mmixslwru2fwa-io'
GOLD_URL = 'https://www.goldapi.io/api/XAU/USD'
SILVER_URL = 'https://www.goldapi.io/api/XAG/USD'

# Belirtilen URL'den fiyat almak için fonksiyon
def get_price(url):
    headers = {
        'x-access-token': API_KEY,  # API anahtarını ekliyoruz
        'Content-Type': 'application/json'
    }
    response = requests.get(url, headers=headers)  # HTTP GET isteği yapılıyor
    if response.status_code == 200: 
        data = response.json() 
        return data['price']  # Altın veya gümüş fiyatı döndürme
    else:
        return "Error"  # Fiyat gelirken hata olursa error yazdırma

# Fiyatları ekranda gösteren fonksiyon
def show_prices():
    clear_ekran()  #Her yeni buton seçildiğinde ekranı temizler diğer buton içeriğinin altına ekleme yapmaması için
    default_color_buton()  # Başka bir butona basıldığında altın butonunu eski rengine çevirir.
    btn_altin.configure(bg="white")  # butona basıldığında rengini beyaz yapar
    
    gold_price = get_price(GOLD_URL)  # Altın fiyatını alır
    silver_price = get_price(SILVER_URL)  # Gümüş fiyatını alır

    if gold_price != "Error":  # Eğer altın fiyatı başarılı bir şekilde alındıysa Label içindeki filtreleri uygulaayarak ekrana yazdırır.
        Label(fr_haberler, text=f"Anlık Altın Fiyatı: {gold_price} USD", anchor='w', font=("Robotobold", 16), bg="#d3d3d3").pack(side=TOP, fill="x")
    else:
        Label(fr_haberler, text="Altın fiyatı alınamadı.", anchor='w', font=("Robotobold", 16), bg="#d3d3d3").pack(side=TOP, fill="x")
    
    if silver_price != "Error":  # Eğer gümüş fiyatı başarılı bir şekilde alındıysa
        Label(fr_haberler, text=f"Anlık Gümüş Fiyatı: {silver_price} USD", anchor='w', font=("Robotobold", 16), bg="#d3d3d3").pack(side=TOP, fill="x")
    else:
        Label(fr_haberler, text="Gümüş fiyatı alınamadı.", anchor='w', font=("Robotobold", 16), bg="#d3d3d3").pack(side=TOP, fill="x")

# Butonların varsayılan rengini ayarlayan fonksiyon
def default_color_buton():
    btn_son_dakika.configure(bg="#BDB76B")
    btn_dunya.configure(bg="#BDB76B")
    btn_ekonomi.configure(bg="#BDB76B")
    btn_sağlık.configure(bg="#BDB76B")
    btn_altin.configure(bg="#BDB76B")

# Ekranı temizleyen fonksiyon
def clear_ekran():
    for widget in fr_haberler.winfo_children():
        widget.destroy()  # Ekrandaki tüm başlık ve linkleri temizler

# URL'yi tıklanabilir hale getiren ve tarayıcıda açan fonksiyon
def open_url(event):
    webview.create_window(event.widget.cget("text"), event.widget.cget("text"))  # Yeni bir pencere oluşturur ve urlyi açar
    webview.start()  # WebView başlatılır

# Haberleri ekrana ekleyen fonksiyon
def add_haberler(haberler):
    haber_count = 0
    for haber in haberler.entries:  # RSS beslemesindeki her bir haber için
        haber_count += 1
        if haber_count > 2:  # İlk iki haberi alır ve sonra döngüden çıkar.
            break
        Label(fr_haberler, text=haber.title, anchor='w', font=("Robotobold", 16), bg="#d3d3d3").pack(side=TOP, fill="x")
        lbl_link = Label(fr_haberler, text=haber.link, anchor='w', font=("Robotobold", 16), bg="#d3d3d3", fg="blue", cursor="hand2")
        lbl_link.pack(side=TOP, fill="x")
        lbl_link.bind("<Button-1>", open_url)  # URL'yi tıklanabilir hale getirir
        canvas = Canvas(fr_haberler, height=10)
        canvas.pack(fill="x")
        canvas.create_line(0, 5, 1000, 5, dash=(8,1), fill="#7C1034", width=2)

# Son dakika haberlerini gösteren fonksiyon
def son_dakika_command():
    clear_ekran()  # Ekranı temizler
    default_color_buton()  # Buton renklerini varsayılan hale getirir
    btn_son_dakika.configure(bg="white")  # son dakika butonunun rengini beyaz yapar
    for url in son_dakika_url:
        haberler = feedparser.parse(url)  # RSS i pars eder
        add_haberler(haberler)  # Haberleri ekrana ekler

# Dünya haberlerini gösteren fonksiyon
def dunya_command():
    clear_ekran()  # Ekranı temizler
    default_color_buton()  # Buton renklerini varsayılan hale getirir
    btn_dunya.configure(bg="white")  # dünya butonunun rengini beyaz yapar
    for url in dunya_url:
        haberler = feedparser.parse(url)  # RSS i pars eder
        add_haberler(haberler)  # Haberleri ekrana ekler

# Ekonomi haberlerini gösteren fonksiyon
def ekonomi_command():
    clear_ekran()  # Ekranı temizler
    default_color_buton()  # Buton renklerini varsayılan hale getirir
    btn_ekonomi.configure(bg="white")  # ekonomi butonunun rengini beyaz yapar
    for url in ekonomi_url:
        haberler = feedparser.parse(url)  #RSS i pars eder
        add_haberler(haberler)  # Haberleri ekrana ekler

# Sağlık haberlerini gösteren fonksiyon
def sağlık_command():
    clear_ekran()  # Ekranı temizler
    default_color_buton()  # Buton renklerini varsayılan hale getirir
    btn_sağlık.configure(bg="white")  # sağlık butonunun rengini beyaz yapar
    for url in sağlık_url:
        haberler = feedparser.parse(url)  # RSS i pars eder
        add_haberler(haberler)  # Haberleri ekrana ekler

# Son dakika haberleri için RSS  url leri
son_dakika_url = ["https://www.ntv.com.tr/son-dakika.rss",
                  "https://www.ensonhaber.com/rss/ensonhaber.xml",
                  "https://www.milliyet.com.tr/rss/rssnew/sondakikarss.xml",
                  "https://www.mynet.com/haber/rss/sondakika"]

# Dünya haberleri için RSS  url leri
dunya_url = ["https://www.ntv.com.tr/dunya.rss",
             "https://www.ensonhaber.com/rss/dunya.xml",
             "https://www.milliyet.com.tr/rss/rssnew/dunyarss.xml",
             "https://www.mynet.com/haber/rss/kategori/dunya/"]

# Ekonomi haberleri için RSS  url leri
ekonomi_url = ["https://www.ntv.com.tr/ekonomi.rss",
               "https://www.ensonhaber.com/rss/ekonomi.xml",
               "https://www.milliyet.com.tr/rss/rssnew/ekonomi.xml",
               "https://www.trthaber.com/ekonomi_articles.rss"]

# Sağlık haberleri için RSS  url leri
sağlık_url = ["https://www.ntv.com.tr/saglik.rss",
              "https://www.ensonhaber.com/rss/saglik.xml",
              "https://www.milliyet.com.tr/rss/rssnew/saglik.xml",
              "https://www.mynet.com/haber/rss/kategori/saglik/"]

# Ana pencere oluşturuluyor
window = Tk()
window.title("Haber Bot Programı")  # Pencere başlığı
window.geometry("1000x800")  # Pencere boyutları

# Haberlerin gösterileceği çerçeve
fr_haberler = Frame(window, height=800)
# Butonların olduğu çerçeve
fr_buttons = Frame(window, relief=RAISED, bg="#7C1034", bd=2)

# Her bir kategori için butonlar oluşturma
btn_son_dakika = Button(fr_buttons, text="Son Dakika", font=("Robotobold", 16), bg="#BDB76B", command=son_dakika_command)
btn_dunya = Button(fr_buttons, text="Dünya", font=("Robotobold", 16), bg="#BDB76B", command=dunya_command)
btn_ekonomi = Button(fr_buttons, text="Ekonomi", font=("Robotobold", 16), bg="#BDB76B", command=ekonomi_command)
btn_sağlık = Button(fr_buttons, text="Sağlık", font=("Robotobold", 16), bg="#BDB76B", command=sağlık_command)
btn_altin = Button(fr_buttons, text="Anlık Altın ve Gümüş", font=("Robotobold", 16), bg="#BDB76B", command=show_prices)

# Butonlar grid şeklinde yerleştirildi yani dikey
btn_son_dakika.grid(row=0, column=0, sticky="ew", padx=5, pady=5)
btn_dunya.grid(row=1, column=0, sticky="ew", padx=5, pady=5)
btn_ekonomi.grid(row=2, column=0, sticky="ew", padx=5, pady=5)
btn_sağlık.grid(row=3, column=0, sticky="ew", padx=5, pady=5)
btn_altin.grid(row=4, column=0, sticky="ew", padx=5, pady=5)

# Çerçeveler grid ile ana pencereye yerleştiriliyor
fr_buttons.grid(row=0, column=0, sticky="ns")
fr_haberler.grid(row=0, column=1, sticky="nsew")

# Pencerenin açık kalmasını sağlar
window.mainloop()
```
